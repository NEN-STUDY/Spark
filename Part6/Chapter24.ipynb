#고급 분석과 머신러닝 개요
- 스파크 고급 분석 도구
  - 데이터 전처리: 데이터 정제 및 피처 엔지니어링
  - 지도 학습
  - 비지도 학습
  - 추천 엔진
  - 그래프 분석
  - 딥러닝

## 고급 분석에 대한 짧은 입문서
- 고급 분석이란, 데이터 기반의 인사이트를 도출하여 핵심 문제를 해결하거나 예측 또는 추천을 하기 위한 기술을 의미
- 머신러닝에서 가장 일반적으로 활용되는 작업
  - 다양한 특징(feature)을 기반으로 각 샘플에 부여된 레이블을 예측하는 분류/회귀 문제를 포함한 지도학습
  - 사용자의 과거 행동에 기반하여 제품을 제안하는 추천엔진
  - 군집 분석, 이상징후 탐지, 토픽 모델링 등 데이터 구조를 파악하기 위한 비지도 학습
  - 소셜 네트워크상에서 유의한 패턴을 찾는 그래프 분석

### 지도 학습
- 레이블(종속변수)을 포함하는 과거 데이터를 사용하여 모델을 학습하고 데이터의 다양한 특징을 기반으로 레이블값을 예측하는 행위
  - 이러한 학습은 경사하강법(gradient descent)과 같은 반복적 최적화 알고리즘으로 진행됨
  - 학습 알고리즘은 기본 모델로 시작하여 반복적으로 진행되는 학습 과정에서 다양한 내부 파라미터를 조정하며 단계적으로 모델 개선
  - ex) 연령(특징)을 기반으로 개인의 소득(종속변수)을 예측
  
- 지도학습은 예측하고자 하는 변수의 타입에 따라 분류와 회귀로 구분함

#### 분류
- 범주형(불연속적이고 유한한 값의 집합) 종속변수를 예측하는 알고리즘을 학습시키는 행위
  - 이진 분류: 두 그룹 중 하나로 분류
    - ex)스팸이냐 아니냐
  - 다중 클래스 분류: 세 가지 이상의 범주로 분류
    - ex)이메일을 4가지 범주로 구분 -> 스팸/개인/업무/기타
    

- 분류 사례
  - 질병 예측
    - 과거 환자의 행동이나 생리학적 특성에 대한 데이터셋을 학습하여 환자가 심장 질환을 앓고 있는지 아닌지 예측
    - 이진 분류: 건강한 심장 / 건강하지 않은 심장
  - 이미지 분류
    - 애플, 구글, 페북 등의 인터넷 기업들은 과거 사진첩에 담긴 사람들의 이미지를 학습에 활용할 수 있음
    - 주어진 사진의 인물이 누구인지 맞추는 문제 등에 적용 가능
  - 고객 이탈 예측
    - 과거 이탈했거나 유지했던 이력이 담긴 고객 데이터를 학습하여 미래 이탈여부를 예측
  - 구매 여부 예측
    - 사용자의 인터넷 검색 패턴, 위치 관련 데이터를 학습하여 웹사이트 방문자의 특정 제품 구매여부를 예측

#### 회귀
- 연속형 변수(실수)를 예측하는 알고리즘을 학습시키는 행위
- 회귀 사례
  - 판매량 예측
    - 과거 판매 데이터를 분석하여 총 제품 판매량 예측
    - 지난주의 판매 데이터를 이용하여 다음날 판매량 예측 가능
  - 신장(키) 예측
    - 부모의 신장을 활용하여 미래 자녀의 신장 예측
  - 관객 수 예측
    - 넷플같은 미디어 기업은 얼마나 많은 가입자가 특정 콘텐츠를 볼 것인지 예측하려고 함

#### 추천
- 고급분석에서 가장 직관적인 응용 영역 중 하나
  - 다양한 제품이나 명시적인 선호도(등급 등)나 암시적인 선호도(관찰된 행동을 통해 도출)를 연구하여 사용자 또는 아이템 간의 유사성을 도출
  - 유사성을 기반으로 특정 사용자와 유사한 사용자가 선호하는 상품을 추천하거나 이미 구매한 상품과 비슷한 상품을 추천
- 추천 사례
  - 영화 추천
    - 넷플은 스파크를 사용한 대규모 영화 추천 서비스를 제공함
      - 사용자가 어떤 영화를 보았고, 어떤 영화를 보지 않았는지 연구하고
      - 해당 영화에 대한 사용자의 평점이 다른 사용자와 얼마나 유사한지 반영함
  - 상품 추천
    - 아마존에서 상품 추천은 매출 증대를 위한 핵심 도구임
      - 고객에게 상품을 추천하기 위해 고객의 장바구니를 구성하는 아이템을 기반으로 과거 장바구니 이력을 탐색하고,
      - 가장 유사한 아이템을 보유한 장바구니에 포함된 다른 아이템을 추천함
      - 또한 각 상품 페이지마다 다른 고객이 구매한 유사 상품을 보여줌

### 비지도 학습
- 주어진 데이터셋에서 특정 패턴을 찾거나 숨겨진 구조적 특징을 발견하는 행위
